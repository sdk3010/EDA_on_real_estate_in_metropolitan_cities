```python
# 1,2,3,4,5,6 bhk flats / apartment
# Cities = "Hyderabad"
  
from selenium import webdriver
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd
import warnings
import requests
warnings.filterwarnings("ignore")

bhk = []
address = []
house_type = []
area = []
doc = []
sqft_p_price = []
price = []
status = []
facing = []
transaction = []
furnishing = []
city = []
l = []
c = ["Hyderabad"]
b = [1,2,3,4,5,6]
for x in b:
#     for i in c:
        url = f"https://www.magicbricks.com/property-for-sale/residential-real-estate?bedroom={x}&proptype=Multistorey-Apartment,Builder-Floor-Apartment,Penthouse,Studio-Apartment&cityName=Hyderabad"
        

        response=requests.get(url)

        response.status_code
#         driver = webdriver.Edge(r"C:\Users\dhipe\Downloads\edgedriver_win64\msedgedriver.exe")
#         driver.get(url)

#         soup = BeautifulSoup(driver.page_source, 'html.parser')

#         driver.close()

        soup=BeautifulSoup(response.text)

        soup.prettify


        title = soup.find_all("h2", class_="mb-srp__card--title")


        for i in title:

            t = i.text
            regex = re.findall("\d+\sBHK",t)

            if regex:
                bhk.append(regex[0])
            else:
                bhk.append(np.nan)


        for i in title:

            t = i.text
            regex = re.findall("in(\s.*)",t)

    #         address.append(regex[0])
            if regex:
                address.append(regex[0])
            else:
                address.append(np.nan) 


        for i in title:

            t = i.text
            regex = re.findall("BHK\s(.*)\sfor",t)

    #         house_type.append(regex[0])
            if regex:
                house_type.append(regex[0])
            else:
                house_type.append(np.nan) 

        title = soup.find_all("div", class_="mb-srp__card__container")


        for i in title:

            t = i.find("div", class_="mb-srp__card__summary__list--item").text
            regex = re.findall("Area(\d+\s\w+)",t)

            if regex:
                area.append(regex[0])
            else:
                area.append(np.nan)

        title = soup.find_all("div", class_="mb-srp__card__estimate")


        for i in title:
            f1=i.find("div",class_="mb-srp__card__price--size")
            
            if f1:
                sqft_p_price.append(f1.text)
            else:
                sqft_p_price.append(np.nan)


        title = soup.find_all("div", class_="mb-srp__card__estimate")


        for i in title:
            t = i.text   

            regex = re.findall("₹(.*\s(?:Lac|Cr))",t)

            if regex:
                        price.append(regex[0])
            else:
                        price.append(np.nan)


        title = soup.find_all("div", class_="mb-srp__card__summary__list")


        for i in title:
            t = i.text

            regex = re.findall("(Under Construction|Ready to Move)",t)
            if regex:
                        status.append(regex[0])
            else:
                        status.append(np.nan)


        features=soup.find_all("div",class_="mb-srp__card__info") 

        for i in features:
            f1=i.find("div",class_="mb-srp__card__summary__list").text
            regex = re.findall("Transaction(New Property|Resale)",f1)


            if regex:
                        transaction.append(regex[0])
            else:
                        transaction.append(np.nan) 


        for i in features:
            f1=i.find("div",class_="mb-srp__card__summary__list").text
            regex = re.findall("Furnishing(Furnished|Unfurnished|Semi-Furnished)",f1)


            if regex:
                        furnishing.append(regex[0])
            else:
                        furnishing.append(np.nan)


d={"Type_of_BHK":bhk, "Address":address, "Type_of_Building":house_type, "Area":area, "price_per_sqft":sqft_p_price, "Price":price,"Status":status,"Transaction":transaction,"Furnishing":furnishing}
da = pd.DataFrame(d)
# print(da)

city.append(c*da.shape[0])

l = city[0]

da['City'] = l
print(da)
```
