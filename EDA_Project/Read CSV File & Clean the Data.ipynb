```python
# Apply the functions to the 'Area' column and replace the values

def sqyrd_to_sqft(area_str):
    area = int(area_str.split()[0])
    return f"{area*9} sqft"
def sqm_to_sqft(area_str):
    area = float(area_str.split()[0])
    return f"{area*10.7639} sqft"

def kottah_to_sqft(area_str):
    area = float(area_str.split()[0])
    return f"{area*720} sqft"


r['Area'] = r['Area'].apply(lambda x: sqyrd_to_sqft(x) if 'sqyrd' in x else x)
r['Area'] = r['Area'].apply(lambda x: sqm_to_sqft(x) if 'sqm' in x else x)
r['Area'] = r['Area'].apply(lambda x: kottah_to_sqft(x) if 'kottah' in x else x)
```

```python
multiplings = {'Lac': 100000, 'Cr': 10000000}

def convert_price(price_str):
    parts = price_str.split()
    value = float(parts[0])
    multiples = multiplings.get(parts[1], 1)
    return value * multiples

r['Price_in_INR'] = r['Price'].apply(convert_price)

print(r)
```
